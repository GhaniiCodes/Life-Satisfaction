```python
import seaborn as sns  
import matplotlib.pyplot as plt   
import pandas as pd   
import numpy as np   
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score
```

```python
Data = pd.read_csv("2023_SouthAsia_MiddleEast_Happiness_Education_Income.csv")
Data = Data.dropna()
```

```python
Data.head()
```

```python
X = Data[["GDP_per_Capita_USD"]]
X.head()
```

```python
Y = Data[["Happiness_Score"]]
Y.head()
```

```python
Data.plot(
    x="GDP_per_Capita_USD",
    y="Happiness_Score",
    grid=True
          )
plt.ylabel("Happiness out of 10")
plt.xlabel("GDP per Capita")
plt.title("GDP vs Happiness")
plt.show()
```

```python
sns.regplot(
    x="GDP_per_Capita_USD",
    y="Happiness_Score",
    data=Data,
    scatter_kws={"color": "blue"},
    line_kws={"color": "red"},
)

plt.xlabel("GDP per Capita")
plt.ylabel("Happiness out of 10")
plt.title("GDP vs Happiness Score")
plt.grid(True)
plt.show()
```

```python
Model = LinearRegression()
```

```python
Model.fit(X,Y)
```

```python
New_X = [[1200]]
```

```python
Model.predict(New_X)
```

```python
New_X_1 = pd.DataFrame({"GDP_per_Capita_USD": [1200,1250,500,1800,2000,1250,1100,2400,1380,1231,990,865,3000]})
```

```python
Prediction = Model.predict(New_X_1)
Prediction
```

```python
sns.regplot(
    x=New_X_1,
    y=Prediction,
    scatter_kws={"color": "blue"},
    line_kws={"color": "red"},
)

plt.xlabel("GDP per Capita")
plt.ylabel("Happiness out of 10")
plt.title("GDP vs Happiness Score")
plt.grid(True)
plt.show()
```

```python
y_true = Data["Happiness_Score"]
y_pred = Model.predict(Data[["GDP_per_Capita_USD"]])

r2 = r2_score(y_true, y_pred)
print("R² Score:", r2)
```
